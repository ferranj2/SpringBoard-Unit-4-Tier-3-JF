#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
#Transposing and resetting index of the DataFrame
properties_T = properties.transpose().reset_index()

#Renaming the columns as the first row.
properties_T.columns = properties_T.iloc[0,:]

#Eliminating the first row as it has now become the column headers.
properties_T = properties_T.drop(0)

#Show a sample of how the data looks like
properties_T.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
#Rename the placeholder values
properties_T.rename(columns = {'Unnamed: 0': 'London Boroughs'}, inplace=True)
properties_T.rename(columns = {pd.NaT: 'Area ID'}, inplace=True)

#Show a sample of how the data looks like
properties_T.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
#Keep the 'London Boroughs' and 'Area ID' column labels. Collapse all time measurements as a single column using melt()
tidy_data = pd.melt(properties_T, id_vars= ['London Boroughs', 'Area ID'])

#Rename the placeholder column names
tidy_data.rename(columns = {0:'Month','value':'Average Prices'},inplace = True)

#Show a sample of how the data looks like
tidy_data.head(5)
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
print('Data types after melt:')
print(tidy_data.dtypes)
print('Corrected Data types:')
tidy_data['Average Prices'] = pd.to_numeric(tidy_data['Average Prices'])
tidy_data['Month'] = pd.to_datetime(tidy_data['Month'])
print(tidy_data.dtypes)
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
#Are there 32 rows in the tidy_data DataFrame? (There's actually 48)
print(tidy_data['London Boroughs'].unique().shape)

#What are some of these rows?
print(tidy_data['London Boroughs'].unique())

#How about missing values?
print(tidy_data.count())

#Drop any and all rows that have null entries:
null_free = tidy_data.dropna(axis = 0,how='any')

#Non-Null entries that are not boroughs: (Taken from Tier 2 of this exercise)
nonBoroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
               'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
               'SOUTH WEST', 'England']

#De-select the non-Borough entries
null_free = null_free[~null_free['London Boroughs'].isin(nonBoroughs)]
null_free.head()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
Barnet_prices = null_free[null_free['London Boroughs'] == 'Barnet']
Barnet_prices.plot(kind = 'line',x='Month',y='Average Prices')
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:
1. look through the `Month` column
2. extract the year from each individual value in that column 
3. store that corresponding year as separate column. 

Whether you go ahead with this is up to you. Just so long as you answer our initial brief: which boroughs of London have seen the greatest house price increase, on average, over the past two decades? 

```python
#Create a list of the names of the boroughs:
boroughs = null_free['London Boroughs'].unique()

#Define a new DataFrame to store price ratio information.
df2 = pd.DataFrame(index = range(33),columns =['Borough','Ratio','Delta$'])

#Running counter to reference df2 entries inside a for loop
k=0

for borough in boroughs:
    #Using Boolean indexing, select all monthly prices in for the current borough in 1998 and 2018
    prices_1998 = null_free[(null_free['London Boroughs'] == borough) & (null_free['Month'].dt.year == 1998)]
    prices_2018 = null_free[(null_free['London Boroughs'] == borough) & (null_free['Month'].dt.year == 2018)]
    
    #Average the monthly prices.
    mean_price_1998 = prices_1998['Average Prices'].mean()
    mean_price_2018 = prices_2018['Average Prices'].mean()
    
    #Compute the ratio as 100%*mean_price_2018/mean_price_1998
    ratio = 100*mean_price_2018/mean_price_1998
    
    #Store the price ratio and borough data back into df2
    df2.loc[k,'Ratio'] = ratio
    df2.loc[k,'Borough'] = borough
    df2.loc[k,'Delta$'] = mean_price_2018-mean_price_1998
    
    #Update Counter
    k = k+1
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

```python
#Sort output information in descending order by ratio.
df2 = df2.sort_values(by='Ratio',ascending = False)

#Output the results
print('Sorted by Ratio of Price increase')
print(df2)

#Plot the ratio data as a bar graph
ax = df2.plot(kind='bar',x='Borough',y='Ratio')
ax.set_ylabel('Percent increase from 1998 to 2018 ')

#Sort output information in descending order by Direct Price increase.
df2 = df2.sort_values(by='Delta$',ascending = False)

#Output the results
print('Sorted by Mean Price increase')
print(df2)

#Plot the Direct price increase data as a bar graph
ax2 = df2.plot(kind='bar',x='Borough',y='Delta$')
ax2.set_ylabel('Mean Price increase from 1998 to 2018 ')
```
